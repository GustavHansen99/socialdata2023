```python
import numpy as np
import pandas as pd
import folium

SF_map = folium.Map([37.7749, -122.4194], zoom_start=13)
folium.Marker([37.77919, -122.41914], popup='City Hall').add_to(SF_map)
SF_map
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data_path = "/Users/Gustav/Downloads/Police_Department_Incident_Reports__Historical_2003_to_May_2018.csv"
data = pd.read_csv(data_path)

# remove data from 2018
data['Date']= pd.to_datetime(data['Date'])
pre_2018_data = data[~(data['Date'] > '2017-12-31')]
```

```python
SF_map_1 = folium.Map([37.7749, -122.4194], zoom_start=13)
map_data_copy = pre_2018_data.copy()
months = ["June", "July"]
map_data = map_data_copy[(map_data_copy["Category"] == "DRUG/NARCOTIC") & (map_data_copy["Date"].dt.month_name().isin(months)) & (map_data_copy["Date"].dt.year == 2016)]
map_data_coords = map_data[["X", "Y"]]
#map_data_coords
for index, coord in map_data_coords.iterrows():
    folium.CircleMarker( location=[ coord[1], coord[0] ], fill_color='#43d9de', radius=4 ).add_to( SF_map_1 )
SF_map_1
```

```python
from folium import plugins
from folium.plugins import HeatMap

SF_heatmap = folium.Map([37.7749, -122.4194], zoom_start=13)

heatmap_data_copy = pre_2018_data.copy()
crime = ["SEX OFFENSES, NON FORCIBLE"]
heatmap_data = heatmap_data_copy[heatmap_data_copy["Category"].isin(crime)]

heat_data = [[row['Y'],row['X']] for _, row in heatmap_data.iterrows()]

HeatMap(heat_data, radius=10, blur=10).add_to(SF_heatmap)
SF_heatmap
```

For the final element of working with heatmaps, let's now use the cool Folium functionality `HeatMapWithTime` to create a visualization of how the patterns of your favorite crime-type changes over time.

```python
from folium.plugins import HeatMapWithTime

df = pre_2018_data.copy()
larceny = df[df["Category"] == "LARCENY/THEFT"]

larceny["Date"] = pd.to_datetime(larceny["Date"])
larceny["Day"] = larceny["Date"].dt.day

september = larceny[(larceny["Date"] >= "2014-09-01") & (larceny["Date"] <= "2014-09-30")]

september = september.sort_values(by=["Day", "Y", "X"])

locations = september.groupby(["Day", "Y", "X"]).size().reset_index(name="count")

data = []

for day in range(1, 31):
    subset = locations[locations["Day"] == day]
    daylist = []
    for i in list(zip(subset["Y"], subset["X"])):
        daylist.append(list(i))
        
    data.append(daylist)

map_sf = folium.Map(location=[37.7749, -122.4194], zoom_start=11.5, tiles = "Stamen Toner")

heatmap = HeatMapWithTime(data,
                          name="Larceny/Theft",
                          auto_play=True,
                          min_opacity=0.8,
                          radius=15,
                          )

heatmap.add_to(map_sf)
#folium.LayerControl().add_to(map_sf)
map_sf
```
